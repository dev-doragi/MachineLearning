### 간질 데이터 읽어서 통계내기

```python
import pandas as pd
import numpy as np
```

```python
eeg_width = 512
eeg_data = []
eeg_result = []

#eeg_file = pd.read_csv('C:/Project/MachineLearning/Data/SALL.csv', header=None)
eeg_file = open('C:/Project/MachineLearning/Data/SALL.csv', "r") 

for line_eeg_data in eeg_file:
    eeg_data.append(abs(float(line_eeg_data.replace('\n',''))))

np_eeg_data = np.array(eeg_data)

for i in range(0, len(np_eeg_data) // eeg_width):
    eeg_width_data = np_eeg_data[(i * eeg_width) : (i + 1) * eeg_width]
    eeg_width_data2 = eeg_width_data ** 2
    eeg_result.append([np.mean(eeg_width_data), np.std(eeg_width_data), np.median(eeg_width_data), 
                       np.mean(eeg_width_data2), np.std(eeg_width_data2), np.median(eeg_width_data2), 1])  # 마지막에 1 추가

print(eeg_width_data)

# DataFrame 생성 시 마지막 열을 'label'로 추가
df4eeg_result = pd.DataFrame(eeg_result, columns=['mean', 'std', 'median', 'mean2', 'std2', 'median2', 'EEG'])

# 결과를 CSV 파일로 저장
df4eeg_result.to_csv('C:/Project/MachineLearning/Result/eeg_result_SALL.csv', index=False, header=True, encoding='cp949')
```

### 간질 환자와 정상 환자 데이터 합쳐서 학습 데이터 만들기

```python
import pandas as pd
from sklearn.utils import shuffle

# SALL과 ZALL 데이터를 읽어옵니다
df_sall = pd.read_csv('C:/Project/MachineLearning/Result/eeg_result_SALL.csv', encoding='cp949')
df_zall = pd.read_csv('C:/Project/MachineLearning/Result/eeg_result_ZALL.csv', encoding='cp949')

df_sall
```

```python
# 두 데이터를 하나로 합칩니다
df_combined = pd.concat([df_sall, df_zall], axis=0)

# 데이터를 랜덤하게 섞습니다
df_combined = shuffle(df_combined).reset_index(drop=True)

df_combined
```

```python
# 합쳐진 데이터를 새로운 CSV 파일로 저장합니다
df_combined.to_csv('C:/Project/MachineLearning/Result/eeg_result_combined.csv', index=False, header=True, encoding='cp949')
```
