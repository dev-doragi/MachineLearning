```python
import pandas as pd
import numpy as np
```

```python
eeg_width = 512
eeg_data = []
eeg_result = []

#eeg_file = pd.read_csv('C:/Project/MachineLearning/Data/SALL.csv', header=None)
eeg_file = open('C:/Project/MachineLearning/Data/SALL.csv', "r") 

for line_eeg_data in eeg_file:
    eeg_data.append(abs(float(line_eeg_data.replace('\n',''))))

np_eeg_data = np.array(eeg_data)

for i in range(0, len(np_eeg_data) // eeg_width):
    eeg_width_data = np_eeg_data[(i * eeg_width) : (i + 1) * eeg_width]
    eeg_width_data2 = eeg_width_data ** 2
    eeg_result.append([np.mean(eeg_width_data), np.median(eeg_width_data), np.mean(eeg_width_data2)])

print(eeg_width_data)

df4eeg_result = pd.DataFrame(eeg_result, columns=['mean', 'median', 'mean2'])
#df4eeg_result = pd.DataFrame(eeg_result)
df4eeg_result.to_csv('C:/Project/MachineLearning/Result/eeg_result_2.csv', index=True, header=True, encoding='cp949')
```

```python
eeg_width= 512
eeg_data = []
eeg_result = []

eeg_file = open('C:/Project/MachineLearning/Data/SALL.csv', "r")
for line_eeg_data in eeg_file:
    eeg_data.append(abs(float(line_eeg_data.replace('\n', ''))))

np_eeg_data = np.array(eeg_data)

for i in range(0, len(np_eeg_data) // eeg_width): # eeg_data를 512로 나눈다. -> 축소
    eeg_width_data = np_eeg_data[(i * eeg_width) : (i+1) * eeg_width]
    eeg_width_data2 = eeg_width_data ** 2
    eeg_result.append([np.mean(eeg_width_data), np.std(eeg_width_data), np.median(eeg_width_data), 
                       np.mean(eeg_width_data2), np.std(eeg_width_data2), np.median(eeg_width_data2)])
print(eeg_width_data)

df4eeg_result = pd.DataFrame(eeg_result, columns=['mean', 'std', 'median', 'mean2', 'std2', 'median2'])
#df4eeg_result = pd.DataFrame(eeg_result)
df4eeg_result.to_csv('C:/Project/MachineLearning/Result/eeg_result.csv', index=True, header=True, encoding='cp949')
```
